```python
from numpy import genfromtxt
my_data = genfromtxt('bv_data.csv', delimiter = ',')
n = len(my_data)-1
x = my_data[range(1,n),0]
y = my_data[range(1,n),1]
```

```python
import matplotlib.pyplot as plt
import pylab
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.4)
n_train = len(X_train)
n_test = len(X_test)
```

```python
from sklearn import cross_validation
kf = cross_validation.KFold(n_train,5)
```

```python
from numpy import linalg
error_train = []
error_cv = []
error_test = []

for d in range(1,10):
    error_sum = 0
    for id1, id2 in kf:
        X1 = X_train[id1]
        y1 = y_train[id1]
        X2 = X_train[id2]
        y2 = y_train[id2]
        coef = np.polyfit(X1, y1, d)
        y2_pred = np.polyval(coef, X2)
        error_sum += linalg.norm(y2_pred-y2)/np.sqrt(len(y2))
        
    coef = np.polyfit(X_train, y_train, d)
    error_train.append(linalg.norm(np.polyval(coef, X_train)-y_train)/np.sqrt(len(X_train)))
    error_cv.append(error_sum/5)
    error_test.append(linalg.norm(np.polyval(coef, X_test)-y_test)/np.sqrt(len(X_test)))
```

```python
plt.plot(error_train)
plt.plot(error_cv)
plt.plot(error_test)
plt.legend(["train","cv","test"])
plt.show()
```
